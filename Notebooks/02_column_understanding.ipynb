## Demographic Features

These features describe basic student characteristics.

- **school**: Type of school attended (GP or MS)
* **sex**: Gender of the student(dtype-string)
    - Data Type: object (string)
    - Measurement Type: Nominal
 
* **age**: Age of the student
    - Data Type: int64
    - Measurement Type: Numerical (Discrete)
- **address**: Urban or rural residence
## Family & Background Features

These features represent family structure, parental education, and socio-economic background.

- **famsize**: Family size (<=3 or >3)
- **Pstatus**: Parents' cohabitation status
- **Medu**: Mother's education level
- **Fedu**: Father's education level
    - Meaning of Values (0–4)
       Value	Meaning
         1) 0:No formal education
         2) 1:Primary education (up to 4th grade)
         3) 2:5th to 9th grade
         4) 3:Secondary education
         5) 4:Higher education (college / university)
- **Mjob**: Mother's occupation
- **Fjob**: Father's occupation
- **guardian**: Student's guardian
- **reason**:
   - values
   - home:School was near to home
   - reputation:- reputed school
   - course: interested_in_course
   - other:- other personal reasons

   - Measurement Type: Ordinal

## Academic Performance (Target Variables)

These variables represent student academic performance and are the main focus of the analysis.

* **G1**: First period grade
   - Data Type: int64  
   - Measurement Type: Numerical (Discrete)
   - **Meaning**:
     Marks obtained in the first term / first internal exam
       * Reflects:
         - Early understanding
         - Initial seriousness of the student
         - Adjustment to the academic year
  
* **G2**: Second period grade
   - Data Type: int64  
   - Measurement Type: Numerical (Discrete)
   - **Meaning**:
     Marks obtained in the second term / mid-year exam
      * Reflects:
        - Consistency
        - Improvement or decline after G1
        - Impact of habits like studytime, absences, internet use
  
* **G3**: Final grade (Primary Target Variable)
   - Data Type: int64  
   - Measurement Type: Numerical (Discrete)
   - **Meaning**:
     Marks obtained in the final exam
     * This is:
         - The most important score
         - The final academic outcome
         - What schools/colleges actually care about
     * This is usually used as:
         - Pass / Fail decision
         - Performance evaluation
         - Promotion eligibility
   * G3 will be treated as the primary performance indicator in later analysis.
