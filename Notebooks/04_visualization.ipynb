```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned_student_data.csv")
df
```

```python
import matplotlib.pyplot as plt
from pyparsing import line

plt.figure(figsize=(20, 12))

plt.plot(df.index, df["G1"], marker="o", )
plt.plot(df.index, df["G2"], marker="^", linestyle="-.")
plt.plot(df.index, df["G3"], marker="s", linestyle="--")

plt.title("Student Performance: G1 vs G2 vs G3", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Marks", fontsize=12)

plt.legend(["G1","G2","G3"])
plt.grid(True)   
plt.show()
```

```python
plt.figure(figsize=(20, 12))

plt.plot(df.index, df["health"], marker="o", )
plt.plot(df.index, df["total_alcohol"], marker="^", linestyle="-.")
plt.plot(df.index, df["G3"], marker="s", linestyle="--")

plt.title("Student Performance: health vs total_alcohol vs G3", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Marks", fontsize=12)

plt.legend(["health","total_alcohol","G3"])
plt.grid(True)   
plt.show()
```

```python
plt.figure(figsize=(20, 12))

plt.plot(df.index, df["famsize"], marker="o", )
plt.plot(df.index, df["health"], marker="^", linestyle="-.")
plt.plot(df.index, df["G3"], marker="s", linestyle="--")

plt.title("Student Performance: famsize vs health vs G3", fontsize=14)
plt.xlabel("Student Index", fontsize=12)
plt.ylabel("Marks/", fontsize=12)

plt.legend(["famsize","health","G3"])
plt.grid(True)   
plt.show()
```

```python
health_map = {
    1: "Very Bad",
    2: "Bad",
    3: "Average",
    4: "Good",
    5: "Happy"
}
df['health_leble'] = df['health'].map(health_map)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,12))
sns.barplot(x="health_leble", y="G3", data=df)

plt.title("Health Condition vs Final Marks (G3)")
plt.xlabel("Health Condition")
plt.ylabel("Final Marks")
plt.grid(True)
plt.show()
```
