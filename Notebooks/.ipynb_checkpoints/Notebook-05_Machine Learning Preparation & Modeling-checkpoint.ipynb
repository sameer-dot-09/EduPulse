## Objective

The objective of this notebook is to prepare the cleaned dataset for machine learning by:
- Selecting relevant features
- Encoding categorical variables
- Splitting data into training and testing sets
- Applying feature scaling
- Training baseline machine learning models to predict final grade (G3)


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned_student_data.csv")
df.head()
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
target="G3"
```

**Targeted variable for our modelling is G3**

```python
features=["G1","G2","studytime","failures","famsup","schoolsup","health","freetime",
          "Dalc","Walc","paid","internet","famsup","schoolsup","activities"]
```

**Assigning variables as X AND Y for features and target from our data set👇🏻**

```python
X=df[features]
Y=df[target]
```

```python
X
```

NOW ENCODING THE BINARY VALUES INTO A NUMERICAL ONE👇🏻

```python
X=pd.get_dummies(X,drop_first=True)
X.head()
```

```python
#train split model
```

TRAIN_TEST_SPLIT FOR OUR DATASET

```python
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
X_TRAIN.head()
```

```python
X_TEST.head()
```

```python
Y_TEST.head()
```

```python
#Scaling the dataframes

scalar=StandardScaler()
X_trained_scaled=scalar.fit_transform(X_TRAIN)
```

```python
X_trained_scaled
```

```python
X_test_scaled=scalar.transform(X_TEST)
```

```python
#FITTING THE MODEL OF LINEAR REGRESSION

LR=LinearRegression()
Model_fit=LR.fit(X_trained_scaled,Y_TRAIN)
y_pred=LR.predict(X_test_scaled)
```

```python
y_pred
```

```python
print("MSE",mse=mean_squared_error(Y_TEST,y_pred))
print("RMSE",rmse=np.sqrt(mse))
print("MAE",mae=mean_absolute_error(Y_TEST,y_pred))
print("Score:",r2=r2_score(Y_TEST,y_pred))
print(r2)
```
