## Objective

The objective of this notebook is to prepare the cleaned dataset for machine learning by:
- Selecting relevant features
- Encoding categorical variables
- Splitting data into training and testing sets
- Applying feature scaling
- Training baseline machine learning models to predict final grade (G3)


```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned_student_data.csv")
df.head()
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,accuracy_score,confusion_matrix

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
target="G3"
```

**Targeted variable for our modelling is G3**

```python
features=["G1","G2","studytime","failures","famsup","schoolsup","health","freetime",
          "Dalc","Walc","paid","internet","famsup","schoolsup","activities"]
```

**Assigning variables as X AND Y for features and target from our data set👇🏻**

```python
X=df[features]
Y=df[target]
```

```python
X
```

NOW ENCODING THE BINARY VALUES INTO A NUMERICAL ONE👇🏻

```python
X=pd.get_dummies(X,drop_first=True)
X.head()
```

## Train-Test Split

The dataset is split into training and testing sets to evaluate model generalization.


```python
X_TRAIN,X_TEST,Y_TRAIN,Y_TEST=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
X_TRAIN.head()
```

```python
X_TEST.head()
```

```python
Y_TEST.head()
```

## Feature Scaling

Numerical features are scaled to ensure fair contribution during model training.


```python
scalar=StandardScaler()
X_trained_scaled=scalar.fit_transform(X_TRAIN)
```

```python
X_trained_scaled
```

```python
X_test_scaled=scalar.transform(X_TEST)
```

## Model 1: Linear Regression

Linear Regression is used as a baseline model to understand linear relationships between features and target.


```python
LR=LinearRegression()
Model_fit=LR.fit(X_trained_scaled,Y_TRAIN)
y_pred=LR.predict(X_test_scaled)
```

```python
y_pred
```

```python
lr_mse=mean_squared_error(Y_TEST,y_pred)
lr_rmse=rmse=np.sqrt(lr_mse)
lr_mae=mean_absolute_error(Y_TEST,y_pred)
r2_lr=r2_score(Y_TEST,y_pred)
print("MSE:",lr_mse)
print("RMSE:",lr_rmse)
print("MAE:",lr_mae)
print("R2 SCORE IS:",r2_lr)
```

## Model 2: Decision Tree Regressor

Decision Trees capture non-linear relationships and feature interactions.


```python
dt=DecisionTreeRegressor(max_depth=5,
    random_state=42)
dt.fit(X_TRAIN,Y_TRAIN)
y_pred_dt=dt.predict(X_TEST)
```

```python
dt_mse=mean_squared_error(Y_TEST,y_pred_dt)
dt_rmse=np.sqrt(dt_mse)
dt_mae=mean_absolute_error(Y_TEST,y_pred_dt)
r2_dt=r2_score(Y_TEST,y_pred_dt)
print("MSE:",dt_mse)
print("RMSE:",dt_rmse)
print("MAE:",dt_mae)
print("R2 SCORE IS:",r2_dt)
```

## Model 3: Random Forest Regressor

Random Forest improves performance by combining multiple decision trees.


```python
rfr=RandomForestRegressor(n_estimators=100,random_state=42,max_depth=8)
model_fitting=rfr.fit(X_TRAIN,Y_TRAIN)
y_pred_rf=model_fitting.predict(X_TEST)
```

```python
print("Random Forest Performance:")
rf_mse=mean_squared_error(Y_TEST,y_pred_rf)
rf_rmse=np.sqrt(rf_mse)
rf_mae=mean_absolute_error(Y_TEST,y_pred_rf)
r2_rf=r2_score(Y_TEST,y_pred_rf)
print("MSE:",rf_mse)
print("RMSE:",rf_rmse)
print("MAE:",rf_mae)
print("R2 SCORE IS:",r2_rf)

```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    RandomForestRegressor(
        n_estimators=200,
        max_depth=8,
        random_state=42
    ),
    X, Y,
    scoring='r2',
    cv=5
)

print("Cross-validated R2 scores:", cv_scores)
print("Mean R2:", cv_scores.mean())

```

Cross-validation was performed to obtain a stable estimate of model performance.
This reduces dependency on a single train-test split.


```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest"],
    "MAE": [lr_mae, dt_mae, rf_mae],
    "RMSE": [lr_rmse,dt_rmse, rf_rmse],
    "R2": [r2_lr, r2_dt, r2_rf]
})

results
```

## Final Model Comparison and Conclusion

Three regression models were trained to predict the final student grade (G3):

- Linear Regression
- Decision Tree Regressor
- Random Forest Regressor

### Model Performance Summary

| Model              | MAE ↓ | RMSE ↓ | R² ↑ |
|-------------------|-------|--------|------|
| Linear Regression | ~1.40 | ~2.19  | ~0.76 |
| Decision Tree     | ~1.39 | ~2.11  | ~0.78 |
| Random Forest     | ~1.52 | ~2.40  | ~0.71 |

### Key Observations

- **Decision Tree Regressor achieved the highest R² score**, indicating the best overall fit on test data.
- **Linear Regression performed competitively**, suggesting a strong linear relationship between predictors (especially G1, G2) and the target variable G3.
- **Random Forest underperformed** compared to expectations due to the relatively small dataset and dominant linear signals from prior grades.

### Important Notes

- Accuracy metric was intentionally excluded as it is not suitable for regression problems.
- Variations in Decision Tree performance are expected due to its high variance nature.
- Prior academic scores (G1, G2) are the strongest predictors of final performance.

### Final Conclusion

The Decision Tree Regressor is selected as the best-performing model for this dataset, while Linear Regression remains a strong and interpretable baseline. The modeling results confirm that historical academic performance is the most influential factor in predicting final grades.
