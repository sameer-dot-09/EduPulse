# Data Preprocessing & Feature Engineering

## Objective
The objective of this notebook is to prepare the dataset for further analysis by:
- Validating data quality
- Checking data types
- Encoding categorical variables
- Creating derived features for better insights



```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw_stud_data.csv")
```

```python
df.isnull().sum()
```

- No missing values were found in the dataset.
- This confirms the dataset is complete and reliable.


```python
df.duplicated().sum()
```

- No duplicate rows were detected.
- Each row represents a unique student record.


```python
df.dtypes
```

- Numerical features are stored as integers (int64).
- Categorical features are stored as object type.
- No incorrect data type conversions were required.


```python
categorical_features = df.select_dtypes(include='object').columns
numerical_features = df.select_dtypes(include='int64').columns
categorical_features, numerical_features
```

- Features were segregated into categorical and numerical types.
- This helps in deciding encoding and scaling strategies later.


```python
binary_map = {
    'yes': 1, 'no': 0,
    'M': 1, 'F': 0,
    'T': 1, 'A': 0
}

df.replace(binary_map, inplace=True)
```

- Binary categorical variables were encoded numerically.
- This enables correlation analysis and machine learning compatibility.


```python
df['total_alcohol'] = df['Dalc'] + df['Walc']
```

- Combined weekday and weekend alcohol consumption
- Helps analyze lifestyle impact on academic performance


```python
df['avg_grade'] = df[['G1', 'G2', 'G3']].mean(axis=1)
```

- Created a holistic(comprehensive assessment of student) academic performance indicator
- Reduces dependency on a single exam score


```python

df['absence_level'] = pd.cut(
    df['absences'],
    bins=[-1, 5, 15, 100],
    labels=['Low', 'Moderate', 'High']
)
```

- Transformed raw absence count into meaningful categories
- Improves interpretability during analysis

- Dataset is now fully validated, encoded, and enhanced
- Ready for Exploratory Data Analysis (EDA)
